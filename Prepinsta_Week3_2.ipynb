### 3.1 Detecting Outliers using IQR method (for Age Column)

```python
q1=df['Age'].quantile(0.25)
q3=df['Age'].quantile(0.75)
iqr=q3-q1
print("Q1: {}, Q3: {}, IQR: {}".format(q1, q3, iqr))
```

```python
upper_limit = q3 + (1.5 * iqr)
lower_limit = q1 - (1.5 * iqr)
print(upper_limit)
print(lower_limit)
```

```python
#Finding the outliers
df.loc[(df['Age'] > upper_limit) | (df['Age'] < lower_limit)]
```

```python
#trimming the data
new_df=df.loc[(df['Age'] < upper_limit) & (df['Age'] > lower_limit)]
print("Before removing the outliers:",len(df))
print("After removing the outliers:",len(new_df))
print("Outliers:",len(df)-len(new_df))
```

```python
sns.boxplot(new_df['Age'])
```

```python
# Capping - changing the outlier values to upper limit (62) and lower limit (14)
new_df = df.copy()  # Use df.copy() instead of df.copy to avoid reference issues

# Update values in the copied DataFrame
new_df.loc[new_df['Age'] > upper_limit, 'Age'] = upper_limit
new_df.loc[new_df['Age'] < lower_limit, 'Age'] = lower_limit

# Update the original DataFrame with the modified values
df['Age'] = new_df['Age']

# Display the DataFrame after capping
print(df)
```

Question: Investigate the range of values in the Rating column. Does it fall within expected rating scales, and how should outliers be treated?


```python
df['Rating']
```

```python
#Investigating the range of values
print("Range of values in 'Rating' column:")
print(df['Rating'].describe())
```

```python
#Assuming the range to be (0 to 5)
lower_bound=0
upper_bound=5
# Counting values less than lower bound
count_lower = (df['Rating'] < lower_bound).sum()
print('count_lower',count_lower)
# Counting values greater than upper bound
count_upper = (df['Rating'] > upper_bound).sum()
print('count_upper',count_upper)
```

```python
# Replacing the  outliers with lower/upper bounds using clip method
new_df=df.copy()
new_df['Rating'] = df['Rating'].clip(lower=lower_bound, upper=upper_bound)
df['Rating'] = new_df['Rating']
#Verifying if the ouliers has been capped
count_lower = (new_df['Rating'] < lower_bound).sum()
print('count_lower',count_lower)
count_upper = (new_df['Rating'] > upper_bound).sum()
print('count_upper',count_upper)
```
